```python
import numpy as np
import pandas as pd

import seaborn as sns

%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
# pakai seaborn untuk buat regression plot

# import dan rapikan data imigrasi Canada
df_can = pd.read_excel("dataset/Canada.xlsx", skiprows=20, sheet_name="Canada by Citizenship", skipfooter=2)

# rapikan data
# hapus kolom2 yang tidak kita pakai
df_can.drop(['AREA','REG','DEV','Type','Coverage', 'OdName', 'AreaName', 'RegName', 'DevName'], axis=1, inplace=True)

# kita tambahkan juga satu kolom untuk total imigrasi ke Canada per tahun
df_can = df_can.T
df_can['Total'] = df_can.sum(axis=1)

# simpan data kolom Total saja
df_total = df_can[['Total']]

# reset index sehingga totalnya jadi kolom, bukan index
df_total.reset_index(inplace=True)

# ganti judul kolom
df_total.columns = ['Year', 'Total']

print(df_total.shape)
df_total.head()
```

```python
# buat regression plot
rp = sns.regplot(x='Year', y='Total', data=df_total)

# tambahin label
rp.set_title("Year vs Total Immigration to Canada")
```

```python
# pilih style plot seaborn
sns.set_style('whitegrid')

# buat regression plot
rp = sns.regplot(x='Year', y='Total', data=df_total)

# tambahin label
rp.set_title("Year vs Total Immigration to Canada")
```
